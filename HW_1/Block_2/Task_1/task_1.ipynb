```python
%%file task_1_script.py

import csv

from mrjob.job import MRJob
from mrjob.step import MRStep


class EpisodeWordCount(MRJob):
    def mapper(self, _, line):
        reader = csv.reader([line], delimiter=' ')
        elements = next(reader)
        character = elements[1]
        
        yield character, 1

    def combiner(self, character, counts):
        yield character, sum(counts)

    def reducer(self, character, counts):
        yield None, (character, sum(counts))

    def reducer_top_characters(self, _, character_counts):
        top_characters = sorted(character_counts, key=lambda x: x[1], reverse=True)[:20]
        for character, count in top_characters:
            yield character, count

    def steps(self):
        return [
            MRStep(mapper=self.mapper, 
                   combiner=self.combiner, 
                   reducer=self.reducer),
            MRStep(reducer=self.reducer_top_characters)
        ]

if __name__ == '__main__':
    EpisodeWordCount.run()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


def plot(path, title):
    with hdfs.open_input_stream(path) as file:
        file_bytes = file.read()
        file_str = file_bytes.decode("utf-8")
        
        lines = file_str.strip().split('\n')
        character_counts = {line.split('\t')[0].strip('"'): int(line.split('\t')[1]) for line in lines}

        plt.figure(figsize=(12, 6))
        sns.barplot(x=list(character_counts.keys()), y=list(character_counts.values()), hue=list(character_counts.keys()), palette="crest")
        plt.title(title)
        plt.xlabel('Character')
        plt.ylabel('Number of replicas')
        plt.xticks(rotation=45, ha='right')
        plt.show()
```
